```python
# Import necessary libraries
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Import our pricing module
from pricing import (
    PricingConfig, 
    TimeValueCalculator, 
    BondPricingAnalyzer,
    OptionsManager,
    GreeksCalculator,
    OptionsStrategies,
    MortgageAnalyzer,
    ComprehensivePricingFramework
)

print("âœ… Successfully imported pricing module components")
```

## 1. Configuration and Setup

First, let's initialize the configuration and main framework.

```python
# Initialize configuration with custom parameters
config = PricingConfig(
    risk_free_rate=0.03,      # 3% risk-free rate
    dividend_yield=0.02,      # 2% dividend yield
    volatility=0.25          # 25% volatility
)

# Create the main pricing framework
pricing_framework = ComprehensivePricingFramework(config)

print(f"Framework initialized with:")
print(f"- Risk-free rate: {config.risk_free_rate:.1%}")
print(f"- Default volatility: {config.volatility:.1%}")
print(f"- Default dividend yield: {config.dividend_yield:.1%}")
```

## 2. Time Value of Money Calculations

Demonstrate basic time value calculations including compound interest, present value, and yield conversions.

```python
# Initialize time value calculator
time_calc = TimeValueCalculator(config)

# Example: Calculate compound interest
principal = 100000  # $100,000
rate = 0.05        # 5% annual rate
periods = 10       # 10 years

# Simple vs Compound Interest
simple_interest = time_calc.simple_interest(principal, rate, periods)
compound_value = time_calc.compound_interest(principal, rate, periods, compounding_frequency=1)

print("ðŸ’° Time Value of Money Examples:")
print(f"Initial Principal: ${principal:,.2f}")
print(f"Simple Interest ({periods} years at {rate:.1%}): ${simple_interest:,.2f}")
print(f"Compound Value ({periods} years at {rate:.1%}): ${compound_value:,.2f}")
print(f"Difference: ${compound_value - simple_interest:,.2f}")

# Yield conversions
annual_rate = 0.06
holding_period_yield = time_calc.annual_to_holding_period_yield(annual_rate, 0.5)  # 6 months
effective_annual_yield = time_calc.nominal_to_effective_yield(annual_rate, 12)  # Monthly compounding

print(f"\nðŸ“Š Yield Conversions (6% nominal rate):")
print(f"6-month holding period yield: {holding_period_yield:.3%}")
print(f"Effective annual yield (monthly compounding): {effective_annual_yield:.3%}")
```

## 3. Bond Pricing and Analysis

Analyze bond pricing, duration, convexity, and yield curve relationships.

```python
# Initialize bond analyzer
bond_analyzer = BondPricingAnalyzer(config)

# Define bond parameters
face_value = 1000      # $1,000 par value
coupon_rate = 0.05     # 5% coupon
years_to_maturity = 10 # 10 years
market_yield = 0.04    # 4% market yield

# Calculate bond metrics
bond_metrics = bond_analyzer.comprehensive_bond_analysis(
    face_value, coupon_rate, years_to_maturity, market_yield
)

print("ðŸ¦ Bond Analysis Results:")
print(f"Bond Price: ${bond_metrics['price']:.2f}")
print(f"Current Yield: {bond_metrics['current_yield']:.3%}")
print(f"Yield to Maturity: {bond_metrics['ytm']:.3%}")
print(f"Modified Duration: {bond_metrics['duration']:.3f} years")
print(f"Convexity: {bond_metrics['convexity']:.3f}")

# Duration and convexity analysis
yield_changes = np.array([-0.02, -0.01, 0, 0.01, 0.02])  # -2% to +2% yield changes
duration_convexity_analysis = bond_analyzer.duration_convexity_analysis(
    face_value, coupon_rate, years_to_maturity, market_yield, yield_changes
)

print(f"\nðŸ“ˆ Price Sensitivity Analysis:")
for i, yield_change in enumerate(yield_changes):
    actual_price = duration_convexity_analysis['actual_prices'][i]
    duration_estimate = duration_convexity_analysis['duration_estimates'][i]
    convexity_estimate = duration_convexity_analysis['convexity_estimates'][i]
    
    print(f"Yield change: {yield_change:+.1%} -> Price: ${actual_price:.2f} ")
    print(f"  Duration estimate: ${duration_estimate:.2f}, Convexity estimate: ${convexity_estimate:.2f}")
```

## 4. Options Pricing Models

Demonstrate various options pricing methods including Black-Scholes, binomial trees, and Monte Carlo simulation.

```python
# Initialize options manager
options_manager = OptionsManager(config)

# Define option parameters
spot_price = 100       # Current stock price
strike_price = 105     # Strike price
time_to_expiry = 0.25  # 3 months (0.25 years)
volatility = 0.20      # 20% volatility
risk_free_rate = 0.03  # 3% risk-free rate

print("ðŸ“Š Options Pricing Comparison:")
print(f"Underlying: ${spot_price}, Strike: ${strike_price}, Time: {time_to_expiry:.2f} years")
print(f"Volatility: {volatility:.1%}, Risk-free rate: {risk_free_rate:.1%}")
print("-" * 60)

# Black-Scholes pricing
bs_call = options_manager.black_scholes(
    spot_price, strike_price, time_to_expiry, risk_free_rate, volatility, 'call'
)
bs_put = options_manager.black_scholes(
    spot_price, strike_price, time_to_expiry, risk_free_rate, volatility, 'put'
)

print(f"Black-Scholes:")
print(f"  Call Price: ${bs_call['price']:.3f}")
print(f"  Put Price:  ${bs_put['price']:.3f}")

# Binomial tree pricing
binomial_call = options_manager.binomial_tree(
    spot_price, strike_price, time_to_expiry, risk_free_rate, volatility, 'call', steps=100
)
binomial_put = options_manager.binomial_tree(
    spot_price, strike_price, time_to_expiry, risk_free_rate, volatility, 'put', steps=100
)

print(f"\nBinomial Tree (100 steps):")
print(f"  Call Price: ${binomial_call['price']:.3f}")
print(f"  Put Price:  ${binomial_put['price']:.3f}")

# Monte Carlo pricing
mc_call = options_manager.monte_carlo_option(
    spot_price, strike_price, time_to_expiry, risk_free_rate, volatility, 'call', simulations=50000
)
mc_put = options_manager.monte_carlo_option(
    spot_price, strike_price, time_to_expiry, risk_free_rate, volatility, 'put', simulations=50000
)

print(f"\nMonte Carlo (50,000 simulations):")
print(f"  Call Price: ${mc_call['price']:.3f} Â± ${mc_call['std_error']:.3f}")
print(f"  Put Price:  ${mc_put['price']:.3f} Â± ${mc_put['std_error']:.3f}")

# Put-Call Parity verification
pcp_result = options_manager.put_call_parity_check(
    bs_call['price'], bs_put['price'], spot_price, strike_price, time_to_expiry, risk_free_rate
)

print(f"\nðŸ”„ Put-Call Parity Check:")
print(f"  Call + PV(Strike) = ${pcp_result['call_plus_pv_strike']:.3f}")
print(f"  Put + Spot = ${pcp_result['put_plus_spot']:.3f}")
print(f"  Difference: ${pcp_result['difference']:.6f} (should be ~0)")
print(f"  Parity holds: {pcp_result['parity_holds']}")
```

## 5. Greeks Calculations

Calculate and analyze option sensitivities (Greeks) for risk management.

```python
# Initialize Greeks calculator
greeks_calc = GreeksCalculator(config)

# Calculate all Greeks for the call option
greeks = greeks_calc.calculate_all_greeks(
    spot_price, strike_price, time_to_expiry, risk_free_rate, volatility, 'call'
)

print("ðŸ” Options Greeks Analysis (Call Option):")
print(f"Delta (Î”):   {greeks['delta']:.4f}    (Price sensitivity to underlying)")
print(f"Gamma (Î“):   {greeks['gamma']:.4f}    (Delta sensitivity to underlying)")
print(f"Theta (Î˜):   {greeks['theta']:.4f}    (Time decay per day)")
print(f"Vega (Î½):    {greeks['vega']:.4f}     (Volatility sensitivity)")
print(f"Rho (Ï):     {greeks['rho']:.4f}      (Interest rate sensitivity)")

# Greeks sensitivity analysis across different spot prices
spot_range = np.linspace(80, 120, 21)  # Stock prices from $80 to $120
greeks_analysis = []

for spot in spot_range:
    greeks_at_spot = greeks_calc.calculate_all_greeks(
        spot, strike_price, time_to_expiry, risk_free_rate, volatility, 'call'
    )
    greeks_analysis.append({
        'spot': spot,
        'delta': greeks_at_spot['delta'],
        'gamma': greeks_at_spot['gamma'],
        'theta': greeks_at_spot['theta']
    })

# Create DataFrame for analysis
greeks_df = pd.DataFrame(greeks_analysis)

print(f"\nðŸ“ˆ Greeks Analysis Summary:")
print(f"Delta range: {greeks_df['delta'].min():.3f} to {greeks_df['delta'].max():.3f}")
print(f"Max Gamma at: ${greeks_df.loc[greeks_df['gamma'].idxmax(), 'spot']:.2f}")
print(f"Max Gamma value: {greeks_df['gamma'].max():.4f}")

# Delta hedging example
portfolio_size = 1000  # 1000 option contracts
hedge_ratio = greeks['delta'] * portfolio_size

print(f"\nðŸ›¡ï¸ Delta Hedging Example:")
print(f"Option position: {portfolio_size} call contracts")
print(f"Delta per contract: {greeks['delta']:.4f}")
print(f"Hedge ratio: {hedge_ratio:.0f} shares to short")
print(f"Portfolio delta after hedging: ~0 (delta neutral)")
```

## 6. Options Strategies

Demonstrate various options trading strategies and their risk/reward profiles.

```python
# Initialize options strategies analyzer
strategies = OptionsStrategies(config)

# Bull Call Spread
print("ðŸ“ˆ Bull Call Spread Strategy:")
bull_spread = strategies.bull_call_spread(
    spot_price, strike_low=100, strike_high=110, 
    time_to_expiry=time_to_expiry, risk_free_rate=risk_free_rate, volatility=volatility
)

print(f"Long call (${bull_spread['long_strike']}): ${bull_spread['long_premium']:.3f}")
print(f"Short call (${bull_spread['short_strike']}): ${bull_spread['short_premium']:.3f}")
print(f"Net premium paid: ${bull_spread['net_premium']:.3f}")
print(f"Max profit: ${bull_spread['max_profit']:.3f}")
print(f"Max loss: ${bull_spread['max_loss']:.3f}")
print(f"Breakeven: ${bull_spread['breakeven']:.3f}")

# Straddle Strategy
print(f"\nðŸŽ¯ Long Straddle Strategy:")
straddle = strategies.long_straddle(
    spot_price, strike_price, time_to_expiry, risk_free_rate, volatility
)

print(f"Call premium: ${straddle['call_premium']:.3f}")
print(f"Put premium: ${straddle['put_premium']:.3f}")
print(f"Total premium: ${straddle['total_premium']:.3f}")
print(f"Upper breakeven: ${straddle['upper_breakeven']:.3f}")
print(f"Lower breakeven: ${straddle['lower_breakeven']:.3f}")
print(f"Profit if stock moves beyond: Â±${straddle['total_premium']:.3f}")

# Iron Condor Strategy
print(f"\nðŸ¦… Iron Condor Strategy:")
iron_condor = strategies.iron_condor(
    spot_price, strike_put_low=90, strike_put_high=95, 
    strike_call_low=105, strike_call_high=110,
    time_to_expiry=time_to_expiry, risk_free_rate=risk_free_rate, volatility=volatility
)

print(f"Net credit received: ${iron_condor['net_credit']:.3f}")
print(f"Max profit: ${iron_condor['max_profit']:.3f}")
print(f"Max loss: ${iron_condor['max_loss']:.3f}")
print(f"Profit range: ${iron_condor['lower_breakeven']:.2f} - ${iron_condor['upper_breakeven']:.2f}")
print(f"Probability of profit: ~{iron_condor.get('prob_profit', 0.6):.1%}")
```

## 8. Comprehensive Analysis Example

Demonstrate the integrated framework for a complete financial analysis.

```python
# Comprehensive portfolio analysis using the main framework
print("ðŸ“Š Comprehensive Financial Analysis:")
print("=" * 50)

# Portfolio parameters
portfolio_value = 1000000  # $1M portfolio
target_return = 0.08       # 8% target return
max_risk = 0.15           # 15% maximum volatility

# Asset allocation example
asset_weights = [0.6, 0.3, 0.1]  # 60% stocks, 30% bonds, 10% alternatives
asset_returns = [0.10, 0.04, 0.12]  # Expected returns
asset_volatilities = [0.18, 0.06, 0.25]  # Volatilities

# Calculate portfolio metrics
portfolio_return = sum(w * r for w, r in zip(asset_weights, asset_returns))
portfolio_volatility = np.sqrt(sum((w * vol) ** 2 for w, vol in zip(asset_weights, asset_volatilities)))

print(f"Portfolio Composition:")
print(f"  Stocks: {asset_weights[0]:.1%} (Expected return: {asset_returns[0]:.1%})")
print(f"  Bonds: {asset_weights[1]:.1%} (Expected return: {asset_returns[1]:.1%})")
print(f"  Alternatives: {asset_weights[2]:.1%} (Expected return: {asset_returns[2]:.1%})")
print(f"\nPortfolio Metrics:")
print(f"  Expected return: {portfolio_return:.2%}")
print(f"  Expected volatility: {portfolio_volatility:.2%}")
print(f"  Sharpe ratio: {(portfolio_return - config.risk_free_rate) / portfolio_volatility:.3f}")

# Risk analysis
print(f"\nðŸš¨ Risk Analysis:")
confidence_level = 0.95
z_score = 1.645  # 95% confidence
daily_var = portfolio_value * portfolio_volatility / np.sqrt(252) * z_score

print(f"Daily VaR (95% confidence): ${daily_var:,.0f}")
print(f"Annual VaR (95% confidence): ${daily_var * np.sqrt(252):,.0f}")
print(f"Maximum drawdown estimate: {portfolio_volatility * 2:.1%}")

# Performance projections
years = 5
projected_value = portfolio_value * (1 + portfolio_return) ** years
worst_case = portfolio_value * (1 + portfolio_return - 2 * portfolio_volatility) ** years
best_case = portfolio_value * (1 + portfolio_return + 2 * portfolio_volatility) ** years

print(f"\nðŸ”® {years}-Year Projections:")
print(f"Expected value: ${projected_value:,.0f}")
print(f"Worst case (2Ïƒ): ${worst_case:,.0f}")
print(f"Best case (2Ïƒ): ${best_case:,.0f}")
print(f"Total expected return: {(projected_value / portfolio_value - 1):.1%}")
```

## 9. Visualization Examples

Create visualizations to better understand the pricing models and relationships.

```python
# Create visualizations for options pricing and Greeks
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')

# Option price vs spot price
spot_range = np.linspace(70, 130, 50)
call_prices = []
put_prices = []
deltas = []

for spot in spot_range:
    call_result = options_manager.black_scholes(
        spot, strike_price, time_to_expiry, risk_free_rate, volatility, 'call'
    )
    put_result = options_manager.black_scholes(
        spot, strike_price, time_to_expiry, risk_free_rate, volatility, 'put'
    )
    greeks_result = greeks_calc.calculate_all_greeks(
        spot, strike_price, time_to_expiry, risk_free_rate, volatility, 'call'
    )
    
    call_prices.append(call_result['price'])
    put_prices.append(put_result['price'])
    deltas.append(greeks_result['delta'])

# Create subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: Option Prices vs Spot Price
ax1.plot(spot_range, call_prices, 'b-', label='Call Price', linewidth=2)
ax1.plot(spot_range, put_prices, 'r-', label='Put Price', linewidth=2)
ax1.axvline(x=strike_price, color='gray', linestyle='--', alpha=0.7, label=f'Strike (${strike_price})')
ax1.set_xlabel('Spot Price ($)')
ax1.set_ylabel('Option Price ($)')
ax1.set_title('Option Prices vs Spot Price')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Delta vs Spot Price
ax2.plot(spot_range, deltas, 'g-', linewidth=2)
ax2.axvline(x=strike_price, color='gray', linestyle='--', alpha=0.7)
ax2.axhline(y=0.5, color='orange', linestyle=':', alpha=0.7, label='Delta = 0.5')
ax2.set_xlabel('Spot Price ($)')
ax2.set_ylabel('Delta')
ax2.set_title('Call Option Delta vs Spot Price')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Bull Call Spread Payoff
payoff_range = np.linspace(80, 120, 100)
spread_payoffs = []

for s in payoff_range:
    # Bull call spread: Long 100 call, Short 110 call
    long_payoff = max(0, s - 100)
    short_payoff = -max(0, s - 110)
    net_payoff = long_payoff + short_payoff - bull_spread['net_premium']
    spread_payoffs.append(net_payoff)

ax3.plot(payoff_range, spread_payoffs, 'purple', linewidth=2)
ax3.axhline(y=0, color='black', linestyle='-', alpha=0.5)
ax3.axvline(x=100, color='blue', linestyle='--', alpha=0.7, label='Long Strike')
ax3.axvline(x=110, color='red', linestyle='--', alpha=0.7, label='Short Strike')
ax3.fill_between(payoff_range, 0, spread_payoffs, where=np.array(spread_payoffs) > 0, 
                alpha=0.3, color='green', label='Profit Zone')
ax3.fill_between(payoff_range, 0, spread_payoffs, where=np.array(spread_payoffs) < 0, 
                alpha=0.3, color='red', label='Loss Zone')
ax3.set_xlabel('Stock Price at Expiration ($)')
ax3.set_ylabel('Profit/Loss ($)')
ax3.set_title('Bull Call Spread Payoff Diagram')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Plot 4: Bond Price vs Yield
yield_range = np.linspace(0.01, 0.08, 50)
bond_prices = []

for y in yield_range:
    price = bond_analyzer.bond_price(face_value, coupon_rate, years_to_maturity, y)
    bond_prices.append(price)

ax4.plot(yield_range * 100, bond_prices, 'navy', linewidth=2)
ax4.axhline(y=face_value, color='gray', linestyle='--', alpha=0.7, label='Par Value')
ax4.axvline(x=coupon_rate * 100, color='orange', linestyle=':', alpha=0.7, label='Coupon Rate')
ax4.set_xlabel('Yield (%)')
ax4.set_ylabel('Bond Price ($)')
ax4.set_title('Bond Price vs Yield (Duration Effect)')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("ðŸ“ˆ Visualizations completed successfully!")
print("\nðŸŽ¯ Key Insights:")
print("â€¢ Call options gain value as spot price increases")
print("â€¢ Delta approaches 1.0 for deep in-the-money calls")
print("â€¢ Bull call spreads limit both profit and loss")
print("â€¢ Bond prices have inverse relationship with yields")
```
